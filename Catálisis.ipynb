# 1. INTRODUCCIÓN A LA CATÁLISIS 

## 1.1. DEFINICIÓN 

La catálisis es un proceso donde se incrementa la velocidad en una reacción química. Este proceso implica el uso de una sustancia que, a pesar de que no es necesaria para que se dé la reacción, esta es parte de la reacción, llamándose catalizador. 

Un catalizador a grandes rasgos es una sustancia que en una reacción aumenta de la velocidad de la reacción, reaccionando, regenerándose y puede ser recuperado al final de la reacción (el catalizador se fragmenta en pequeñas partículas para acelerar el proceso). Esta formulación habla sobre el comportamiento cinético distintivo del catalizador frente a los reactivos dando un mecanismo diferente a la reacción. Que ocurre a mayor velocidad, en ambas direcciones, ya sea una rección directa o inversa. 

Podemos decir que un catalizador aumenta la velocidad cinética de una reacción sin alterar sin alterar la variación global de energía de Gibbs (ΔG°) del proceso, y sin consumirse durante el mismo. Este no debe ser confundido con un iniciador ya que el iniciador “activa” la reacción, pero se consume totalmente en el proceso. El catalizador tampoco es capaz de intercambiar energía con el sistema ni realiza trabajo sobre el sistema de reacción, por ende, no se altera el equilibrio termodinámico en el proceso ya que su trabajo es solo es cambiar la velocidad de reacción, entonces no modifica las funciones del sistema, como la energía libre de Gibbs (ΔG), la entalpía (ΔH) o la entropía (ΔS), ni influye en el valor de la constante de equilibrio.

## 1.2. CLASES DE CATALIZADORES 





### 1.2.1. CATALIZADORES HOMOGENEOS:
 Un catalizador homogéneo es aquel que está en el mismo estado de los reactivos, normalmente en estado liquido y gaseoso. Esto tiene una gran ventaja, ya que están en un contacto estrecho, esto permite una interacción molecular significativa, pero esto también es una característica en contra ya que hace difícil aislar el catalizador de la reacción.


![Tomado de :https://es.paperblog.com/catalisis-heterogenea-1926248/4](catalisis.jpeg)

### 1.2.2. CATALIZADORES HETEROGENEOS:
 Un catalizador heterogéneo es uno que funciona en una fase, normalmente es sólido e interactúa con los reactivos en fase liquida o gaseosa. Los sitios activos en los catalizadores adsorben reactivos, debilitando enlaces y facilitando reacciones.

![Tomado de :https://www.scielo.org.mx/scielo.php?script=sci_arttext&pid=S2448-56912021000100004](CATAHOMO.png)

### 1.2.3. BIOCATALISIS:
 La biocatálisis o catálisis enzimática, adopta el uso de compuestos biológicos para poder catalizar reacciones químicas. Esta hace fácil un amplio rango de reacciones enfocadas en el carbono que tienen lugar en entornos sin células, completamente in vitro, hasta procesos con cultivos de células vivas

## 1.3. PROPIEDADES DE LOS CATALIZADORES
Todo catalizador tiene ciertas propiedades que afectan su desempeño, estas propiedades fisicas y quimicas son son:

Propiedades fisicas: 

- Tamaño de particula: Es el diametro nomimal de cada elemento individual del catalizador 
- superficie de particula: Es el area disponible para la reacción, y esta abarca ademas de la superficie, el interior de los poros 
- volumen de los poros: Es la cantidad de masa libre al interior de los poros 
- Tamaño de los poros: Magnitud para expresar la anchura de los poros
- Estabilidad: la capacidad de soportar condiciones de la reacción, como las altas temperaturas, la presión o exposición a ciertas sustancias químicas sin degradarse 
- Regenerabilidad: Los catalizadores pueden regenerarse o reutilizarse, para ser rentables 


![Tomado de :https://espanol.noahchemicals.com/blog/los-cinco-catalizadores-quimicos-comunes-usados-en-la-fabricacion/](fotofis.jpeg)

Propiedades quimicas:

- Composición quimica: Es la mezcla de diferentes elementos que componen el catalizador 
- Estructura quimica: Agrupamiento geometrico, propiedades y enlaces de los atomos presentes 
- Concentración de centros activos: Es la fracción de la fracción de la superficie de los catalizadores  con propiedades catalicas
- Acidez: Capacidad de ceder protones o capturar electrones de las especies activas
- Selectividad: los catalizadores pueden favorecer algunas reacciones sobre otra, dando la posibilidad de formar un producto con lo mínimo de subproductos

## 1.4. TIPOS CATÁLISIS 
### 1.4.1. CATÁLISIS ÁCIDA 

La catálisis ácida es un proceso donde una reacción se ve acelerada por un ácido. Esta puede ser ampliamente usada, ya que el ácido que actúa como catalizador  no se consume y puede hacer que algunos reactivos sigan existiendo en los productos 

![Tomado de : https://www.wikiwand.com/es/articles/Cat%C3%A1lisis_%C3%A1cida_de_Lewis](images.jpg)

### 1.4.2 CATÁLISIS BÁSICA

En este proceso la catálisis la realiza una base. Estas proporcionan electrones negativos para formar enlaces con otras moléculas y son cruciales en la industria química .

![Tomado de : https://www.ingenieriaquimicareviews.com/2020/08/fundamentos-de-catalizadores-heterogeneos.html](fundamentos.jpg)

### 1.4.3 CATALISIS METALICA

Los catalizadores metálicos aceleran la reacción sin consumirse en el proceso, estos tiene una alta superficie especifica que permite muchos más sitios activos en la reacción, tienen un gran estabilidad térmica y resistencia a la oxidación.

![Tomado de : https://www.ehu.eus/es/web/metal-catalysis-and-organocatalysis](images1.jpg)

### 1.4.4. CATALISIS ENZIMATICA

Este proceso ampliamente usado en la bioquímica permite entender como las enzimas aceleran procesos químicos en organismos. Estás proteínas actúan como catalizadores formando un complejo entre la encima y el sustrato, luego se transforma el sustrato en productos ya que la enzima facilita la formación de estos.

# 6. APLICACIONES DE LA CATÁLISIS


## 6.1 APLICACIONES INDUSTRIALES DE LA CATÁLISIS HOMOGENEA Y HETEROGENEA  

 En la catálisis se encuentra la facilitación de toda la industria química, ya sean refinerías o farmacéuticas, apareciendo en casi cualquier producto, como los combustibles fósiles o biomasas, dando así la creencia de que mas del 90% de los productos de la industria tuvieron la intervención de un catalizador. En cualquier proceso industrial donde esté presente la catálisis, esta cumplirá la función de hacer mucho más fácil dicho proceso, ya sea en la reducción de tiempo del proceso, la mejora de rendimiento o la disminución de costos. Ya en planta la selección del catalizador o catalizadores correctos es un proceso exhaustivo que requiere de una amplia investigación, en especial en procesos al granel. El costo del catalizador debe reducirse lo máximo posible, normalmente aumentando las cifras de rotación.  

 


### 6.1.1. APLICACIONES INDUSTRIALES DE LA CATALISIS HOMOGENEA

 En la catálisis homogénea se puede encontrar una de las partes más importantes dentro de los procesos industriales modernos, específicamente en aquellos de selectividad fina y el control molecular, como pasa en la producción farmacéutica o en la producción de químicos de alto valor.  En estas aplicaciones la catálisis homogénea transforma sustratos complejos con mucha precisión, dando mejoras en la eficiencia, en la reducción de subproductos y pasos de purificación.
 A nivel global la cantidad de procesos industriales que dependen de la catálisis homogénea es enorme, muchos de estos basados en complejos metálicos, como en la hidroformilación, que permite la formación de aldehídos a gran escala y con una altísima selectividad, y esto a su vez, impacta la producción de detergentes, plásticos y solventes. Las reacciones de polimerización controlada gracias a la catálisis se han podido abrir rutas industriales que eran consideradas, prácticamente inviables.
 En la selección en planta, el catalizador homogéneo requerido se escoge respecto a su estabilidad, su tolerancia a impurezas y su capacidad de recuperarse o reciclarse, siendo esta su facultad más importante, ya que el costo de los metales de transición utilizados es muy alto. Por eso en la actualidad se busca nuevas estrategias como lo son los disolventes ionios, membranas selectivas o sistemas bifásicos acuosos que permitan acuosos, que permitan conservar la selectividad de la catálisis homogénea, pero con costos de utilización rentables. Estos avances científicos en los últimos años han permitido usar de manera industrial la catálisis homogénea que es esencial para la química contemporánea.

### 6.1.2.  APLICACIONES INDUSTRIALES DE LA CATALISIS HETEROGENEA 

La catálisis heterogénea es prácticamente uno de los motores fundamentales en toda la industria química actual. Estos procesos son utilizados ampliamente en refinerías, plantas petroquímicas y unidades de conversión energética, apareciendo en el día a día de las personas, ya sea en combustibles, fertilizantes, plásticos, o incluso materiales muchísimo mas avanzados. 
Sim la catálisis heterogénea la gran mayoría de productos serían imposible de hacer de una manera viable ya que todos estos dependen de los catalizadores sólidos. El fin de usar este tipo de catalizadores es reducir los costos operativos y el costo energético. Por ejemplo, en el reformado catalítico y el craqueo en refinerías, los catalizadores solidos permiten transformar grandes cantidades de petróleo en sus derivados de alto valor.
En la selección del catalizador se debe tener en cuenta que el catalizador no solo debe ser funcional al inicio del proceso, sino que pueda soportar gran cantidad de horas sin desactivarse, por ende, actualmente se busca la estabilidad, la dispersión metálica controlada y resistencia a la sinterización o al envenenamiento de impurezas. Este tipo de catálisis es la base de energías emergentes como la transformación de la biomasa la utilización del CO₂ y la producción de hidrogeno. Todos estos factores convierten la catálisis heterogénea en la base de la industria actual y del futuro.
