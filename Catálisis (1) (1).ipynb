# 1. INTRODUCCIÓN A LA CATÁLISIS 

## 1.1. DEFINICIÓN 

La catálisis es un proceso donde se incrementa la velocidad en una reacción química. Este proceso implica el uso de una sustancia que a pesar de que no es necesaria para que se dé la reacción, esta es parte de la reacción, llamándose catalizador. 

Un catalizador a grandes rasgos es una sustancia que aumenta de la velocidad deL proceso, reaccionando, regenerándose y puede ser recuperado al final de la reacción (el catalizador se fragmenta en pequeñas partículas para acelerar el proceso). Esta formulación habla sobre el comportamiento cinético distintivo del catalizador frente a los reactivos dando un mecanismo diferente a la reacción. Que ocurre a mayor velocidad, en ambas direcciones, ya sea una rección directa o inversa. 

Podemos decir que un catalizador aumenta la velocidad cinética de una reacción sin alterar la variación global de energía de Gibbs (ΔG°) del proceso, y sin consumirse durante el mismo. Este no debe ser confundido con un iniciador, ya que el iniciador “activa” la reacción, pero se consume totalmente en el proceso. El catalizador tampoco es capaz de intercambiar energía con el sistema ni realiza trabajo sobre el sistema de reacción, por ende, no se altera el equilibrio termodinámico en el proceso ya que su trabajo es solo es cambiar la velocidad de reacción, entonces no modifica las funciones del sistema, como la energía libre de Gibbs (ΔG), la entalpía (ΔH) o la entropía (ΔS), ni influye en el valor de la constante de equilibrio.

# 2. BASES PRELIMINARES

## 2. Principios de termodinámica aplicados a la catálisis
En un sistema catalítico, la termodinámica determina si una reacción es espontánea, mientras que el catalizador sólo modifica la ruta por la que ocurre y la condición de espontaneidad viene dada por la variación de energía libre de Gibbs. En catálisis heterogénea, además de la termodinámica química se deben considerar los fenómenos de transferencia de calor y materia entre fase fluida y superficie catalítica, estos fenómenos modifican la “velocidad observada” y pueden ocultar la cinética intrínseca si existen resistencias de transporte significativas.


## 2.1 Equilibrio químico en presencia de catalizadores
El equilibrio químico se caracteriza por una constante K que depende únicamente de la temperatura y de la variación de energía libre estándar.
La literatura de equilibrio y catálisis coincide en que un catalizador no altera el valor de K; sólo acelera por igual las velocidades directa e inversa. Por tanto, propiedades termodinámicas como rendimientos máximos, selectividades en equilibrio o composición de fase no se modifican introduciendo un catalizador, aunque en la práctica sí puede cambiar la distribución de productos bajo condiciones de operación limitadas por cinética o por transferencia de masa.


## 2.2 Principio de Le Chatelier aplicado a reacciones catalizadas
El principio de Le Chatelier establece que, si un sistema en equilibrio se somete a un cambio de concentración, presión o temperatura, el sistema se desplaza en la dirección que reduzca ese “estrés”.
Los textos de química (referenciados) subrayan que la adición de un catalizador no es un estrés termodinámico; no modifica presiones ni concentraciones de equilibrio, de modo que Le Chatelier no predice ningún desplazamiento del equilibrio al introducir un catalizador.
No obstante, en procesos industriales trabajados lejos del equilibrio (por limitaciones de tiempo de residencia o transferencia de masa), el catalizador puede favorecer rutas preferenciales, lo que se traduce en cambios aparentes en conversión o selectividad a escala de reactor.

## 2.3 Factores que afectan la actividad catalítica
Además de la naturaleza química del catalizador, su actividad está fuertemente influida por:
1.	Temperatura.
La dependencia Arrhenius hace que la velocidad intrínseca aumente exponencialmente con la temperatura, pero temperaturas muy elevadas pueden promover desactivación por sintonización o cambios de fase.
2.	Presión y concentración.
En reacciones gas–sólido, cambios de presión modifican las coberturas de adsorbatos y la velocidad de reacción. Revisiones sobre transferencia de masa muestran que la composición y el caudal influyen en el coeficiente global de transferencia, pudiendo dar lugar a limitaciones difusionales que enmascaran la cinética intrínseca.
3.	Transferencia de masa y calor.
Textos sobre catálisis heterogénea señalan que resistencias externas (película fluida) e internas (difusión en poros) pueden reducir la velocidad observada y disminuir la aparente energía de activación, haciendo que el proceso parezca menos sensible a la temperatura.
4.	Estructura y morfología del catalizador.
El tamaño de poro, la distribución de fases y la dispersión del metal afectan la accesibilidad y naturaleza de los sitios activos. Un exceso de microporos puede generar limitaciones difusionales, mientras que mesoporos adecuados favorecen la transferencia de masa.
5.	Condiciones hidrodinámicas.
El coeficiente de transferencia de masa aumenta con la agitación o velocidad del fluido; estudios en sistemas gas-líquido muestran una fuerte dependencia de la constante volumétrica de transferencia con la temperatura y el régimen de flujo.


# 3. Mecanismos de acción catalítica
Las enzimas son catalizadores selectivos, son capaces de reconocer entre miles de moléculas presentes en un medio a un único sustrato específico.
## 3.1 Key-Lock
En 1894, el químico Emil Fischer propuso el modelo Key-Lock para intentar explicar esta capacidad de las enzimas de elegir el reactivo correcto. Fischer planteaba que las enzimas presentan un alto grado de especificidad hacia el sustrato, por lo que se asume que el sitio activo de la enzima y el sustrato encajan perfectamente entre sí, así cada uno tiene formas y geometrías complementarios predeterminados. Entonces la especificidad se puede explicar de forma análoga a una llave y cerradura, donde la cerradura es la enzima y la llave es el sustrato.

Primero se produce la unión del sustrato a la enzima en su sitio activo hasta formarse el complejo enzima-sustrato; las enzimas catalizan la reacción química y como resultado se produce la formación de uno o más productos y se liberan las enzimas para reusarse en la siguiente reacción. Veamos la siguiente imagen para ilustrar mejor el modelo 

![Tomado de: https://microbenotes.com/lock-and-key-model-mode-of-action-of-enzymes/](keylock.JPg)

Este modelo tiene ciertas limitaciones, por ejemplo no llega a explicar cómo se estabiliza el complejo enzima-sustrato en el estado de transición, además no descrube la unión de múltiples sustratos a la enzima y afirma que la enzima posee estructura rígida  que no cambia al unirse al sustrato, sin embargo, recientes inestigaciones muestran que si se produce un cambio en el sitio activo de la enzima al unirse al sustrato.
## 3.2 Sitio inducido 
En 1959, Daniel Koshland propuso un cambio en el modelo de Key-lock, este cambio plantea que el sitio activo de una enzima puede ser más flexible, de tal forma que el sitio activo puede modificar su forma a medida que el sustrato interactúa con la enzima.
Por ejemplo, una analogía adecuada para describir el modelo de ajuste inducido es la de una mano modificando la forma de un guante mientras la persona se lo pone. En este caso, el guante es el sitio activo de la enzima y la mano es el sustrato.

# 4. Modelos Cinéticos de Catálisis
## 4.1 Cinética química
La cinética química se encarga de estudiar la rapidez con la que se producen las reacciones químicas y los factores que la afectan, es decir, determina cómo es el cambio de la concentración de reactantes y productos con respecto al tiempo. Como vimos en las secciones anteriores el objetivo de los catalizadores es modificar la velocidad de las reacciones, por lo que es importante estudiar la cinética química cuando hay presencia de un catalizador.

### 4.1.1 Ley de Velocidad y Órdenes de Reacción en Sistemas Catalizados
Recordemos que la Ley de velocidad decribe cómo es la relación entre la velocidad de una reacción química y la concentración de sus reactivos. Para una reacción:
$$
aA + bB \rightarrow productos
$$
donde a y b son los coeficientes estequiométricos, entonces la Ley de velocidades para la reacción se expresa mediante la siguiente ecuación:
$$
v = k[A]^m[B]^n
$$

En esta ecuación $[A]$ y $[B]$ son las concentraciones molares de los reactivos, $k$ es la constante de velocidad para la reacción a una temperatura especifica y m y n son los órdenes de reacción que generalmente son enteros positivos. 

En este contexto lo es de nuestro interes es conocer cómo cambia la Ley de Velocidad cuando interviene un catalizador, a continuación un ejemplo para ilustrar mejor:


Supongamos que la reacción:

$$
A \rightarrow P
$$

puede ocurrir tanto sin catalizador como con catalizador, veamos como son las velocidades en ambos casos para una misma concentración inicial de $A$.


Cuando no hay catalizador la reacción sigue una ley de primer orden:

$$
v_{\text{no cat}} = k_0[A]
$$

donde

$k_0 = 0.10 \, \text{s}^{-1}$



En presencia de un catalizador $C$, la reacción sigue la ley:

$$
v_{\text{cat}} = k[A][C]
$$

donde para este ejemplo:

$k = 0.50 \, \text{L*mol}^{-1}\text{s}^{-1}$
Concentración del catalizador:
  $$
  [C] = 1.0 \, \text{mol* L}^{-1}
  $$

La concentración inicial del reactivo

$$
[A] = 0.10 \, \text{mol*L}^{-1}
$$

Entonces la velocidad sin catalizador

$$
v_{\text{no cat}} = k_0[A]
$$

$$
v_{\text{no cat}} = (0.10)(0.10) = 0.010 
\, \text{mol*L}^{-1}\text{s}^{-1}
$$

Y la velocidad con catalizador

$$
v_{\text{cat}} = k[A][C]
$$

$$
v_{\text{cat}} = (0.50)(0.10)(1.0)
$$

$$
v_{\text{cat}} = 0.050 
\, \text{mol*L}^{-1}\text{s}^{-1}
$$

Veamos una breve comparación

| Caso             | Ecuación          | Velocidad $(\text{mol *L}^{-1}\text{s}^{-1})$ |
|------------------|-------------------|-----------------------------------------------|
| Sin catalizador  | $v = k_0[A]$    | **0.010**                                     |
| Con catalizador  | $v = k[A][C]$  | **0.050**                                     |


Es evidente que:
- La velocidad **aumenta 5 veces** cuando se añade el catalizador.  
- La reacción catalizada depende tanto de $[A]$ como de $[C]$.  
- Esto muestra que el catalizador no solo acelera la reacción, sino que puede modificar la forma de la Ley de Velocidad, introduciendo una nueva ruta con una energía de activación más baja.

Veamos cómo sería la gráfica:

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Parámetros de la reacción del ejemplo ----
k_no_cat = 0.10      # constante sin catalizador
k = 0.50             # constante de la ruta catalizada
C = 1.0              # concentración del catalizador (fija)

# ---- Rango de concentraciones de A ----
A = np.linspace(0, 0.2, 300)   # el ejemplo usa [A]=0.10 mol/L, pero graficamos un rango

# ---- Leyes de velocidad ----
v_no_cat = k_no_cat * A
v_cat    = k * A * C


plt.figure(figsize=(8,5))
plt.plot(A, v_no_cat, label="Sin catalizador (k = 0.10)", linewidth=3)
plt.plot(A, v_cat, label="Con catalizador (k = 0.50, C = 1.0)", linewidth=3)

plt.title("Comparación de velocidades: reacción catalizada vs no catalizada")
plt.xlabel("Concentración de A (mol/L)")
plt.ylabel("Velocidad v (mol/L·s)")
plt.grid(True)
plt.legend()
plt.show()
```

## 4.2 Modelo de Michaelis-Menten

Este modelo desarrollado por Leonor Michaelis y Maud Menten en 1913 describe cómo evoluciona la v3elocidad de una reacción enzimática en función de la concentración del sustrato; este modelo cinematico parte del mecanismo básico en el que la enzima $E$ se une reversiblemente al sustrato $S$ para formar un complejo intermedio $ES$, que posteriormente se transforma en el producto $P$ y la enzima queda libre otra vez, tanto $E$ como $S$ están en la misma fase, normalmente en solución acuosa, por lo tanto es un caso particular de catálisis homogenea. Este proceso se describe mediante la relación:
$$
E + S \xrightleftharpoons[k_{-1}]{k_1} ES \xrightarrow{k_2} P
$$
La velocidad global de formación de producto está dado por:
$$
\frac{d[P]}{dt} = r = k_2[ES] 
$$
Pero como $[ES]$ es un intermedio, se usa la aproximación de estado estqcionario:
$$
\text{Tasa de formación de } [ES] = \text{Tasa de descomposición de } [ES] 
$$
La tasa de formación de \([ES]\) es:

$$
k_1[E][S]
$$

La tasa de descomposición de \([ES]\) es:

$$
k_{-1}[ES] + k_2[ES] 
$$

Igualando ambas tasas:

$$
k_1[E][S] = (k_{-1} + k_2)[ES] 
$$

De donde se obtiene:

$$
[ES] = \frac{k_1[E][S]}{k_{-1} + k_2}
$$

Luego, Michaelis y Menten definieron una constante:
$$
K_m = \frac{k_{-1} + k_2}{k_1} 
$$

Reemplazando:
$$
[ES] = \frac{[E][S]}{K_m} 
$$

La concentración total de enzima se expresa como:

$$
[E_0] = [E] + [ES] 
$$

Donde \([E]\) es la concentración de enzima libre (no unida) y \([ES]\) es la concentración de enzima unida al sustrato.  


$$
[E] = [E_0] - [ES] 
$$

Sustituyendo el valor de \([E]\), se obtiene:

$$
[ES] = \frac{([E_0] - [ES])[S]}{K_m} 
$$

Que también puede escribirse como:

$$
[ES] = \frac{[E_0][S] - [ES][S]}{K_m}
$$

Reordenando nuevamente para despejar \([ES]\):

$$
K_m[ES] = [E_0][S] - [ES][S] 
$$

$$
K_m[ES] + [ES][S] = [E_0][S] 
$$

$$
[ES] = \frac{[E_0][S]}{K_m + [S]}
$$

Sustituyendo este valor de \([ES]\), obtenemos:

$$
v = \frac{k_2 [E_0][S]}{K_m + [S]} 
$$

Si la velocidad de la reacción catalizada por la enzima se mide en la etapa muy inicial, entonces la concentración de sustrato puede reemplazarse por la concentración inicial del mismo, \([S_0]\).  
Por lo tanto, la velocidad inicial de la reacción enzimática (\(r_0\)) es:

$$
r_0 = \frac{k_2 [E_0][S_0]}{K_m + [S_0]} 
$$

La ecuación anterior es la conocida **ecuación de Michaelis–Menten**.
## Orden de las reacciones catalizadas por enzimas

Para saber cómo es el orden de esta ecuación, veamos los siguientes esenarios


### 1) A alta concentración de sustrato \([S_0]\)

Cuando la concentración inicial de sustrato es muy alta, \(K_m\) puede despreciarse frente a \([S_0]\), y la ecuación toma la forma:

$$
r_0 = \frac{k_2 [E_0][S_0]}{[S_0]} = k_2 [E_0] 
$$
Por lo tanto, la reacción será de orden cero respecto a la concentración de sustrato y de primer orden respecto a la concentración total de enzima**.


### 2) A baja concentración de sustrato \([S_0]\)

Cuando la concentración inicial de sustrato es baja, \([S_0]\) puede considerarse pequeña frente a \(K_m\), y la ecuación toma la forma:

$$
r_0 = \frac{k_2}{K_m} [E_0][S_0] 
$$

Por lo tanto, la reacción será de primer orden respecto a la concentración de sustrato y también de primer orden respecto a la concentración total de enzima.




```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Vmax = 1.0
Km = 0.8

# Rango de sustrato
S = np.linspace(0, 5, 400)

# Michaelis-Menten
r = Vmax * S / (Km + S)


plt.figure(figsize=(8,6))


plt.plot(S, r, color="black", linewidth=2.8)


plt.axhline(Vmax, color="gray", linestyle="--", linewidth=1, alpha=0.7)
plt.axhline(Vmax/2, color="gray", linestyle="--", linewidth=1, alpha=0.7)
plt.axvline(Km, color="gray", linestyle="--", linewidth=1, alpha=0.7)


plt.text(Km, -0.07, r"$K_m$", ha="center", fontsize=13)
plt.text(0.15, Vmax + 0.03, r"$r_{\max}$", fontsize=13)
plt.text(0.15, Vmax/2 + 0.03, r"$\frac{r_{\max}}{2}$", fontsize=13)


plt.text(0.9, 0.4, "Primer orden\nen [S₀]", fontsize=13, alpha=0.85)
plt.text(3.2, 0.92, "Orden cero\nen [S₀]", fontsize=13, alpha=0.85)


plt.xlabel("Concentración de sustrato [S₀]", fontsize=13)
plt.ylabel("Velocidad inicial $r_0$", fontsize=13)


plt.xlim(0, 5)
plt.ylim(0, 1.15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.title("Cinética de Michaelis–Menten\nTransición de primer orden a orden cero", fontsize=15, pad=15)

plt.grid(alpha=0.15)

plt.show()
```

## 4.3 Lineweaver-Burk
Es una transformación lineal de la ecuación de Michaelis-Menten
Partimos de la ecuación de Michaelis–Menten:

$$
v = \frac{V_{\max}[S]}{K_m + [S]}
$$

Tomando recíprocos en ambos lados:

$$
\frac{1}{v} = \frac{K_m}{V_{\max}} \cdot \frac{1}{[S]} 
+ \frac{1}{V_{\max}}
$$

La forma anterior es equivalente a la ecuación de una recta:

$$
y = mx + b
$$

donde:

$$
y = \frac{1}{v}
$$
$$
x = \frac{1}{[S]}
$$
$$
m = \frac{K_m}{V_{\max}} \text{(pendiente)}
$$
$$
b = \frac{1}{V_{\max}}  \text{(intercepto en el eje y)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la enzima
Vmax = 1.0
Km = 0.8

# Rango de sustrato
S = np.linspace(0.1, 5, 100)  # evitar cero para no dividir por 0

# Michaelis-Menten
v = (Vmax * S) / (Km + S)

# Lineweaver-Burke
inv_S = 1 / S
inv_v = 1 / v


fig, ax = plt.subplots(1, 2, figsize=(13,5))

# ---- Gráfica Michaelis–Menten ----
ax[0].plot(S, v, color="black", linewidth=2.5)
ax[0].set_title("Michaelis–Menten", fontsize=14)
ax[0].set_xlabel("[S] (mol/L)", fontsize=12)
ax[0].set_ylabel("v (mol/L·s)", fontsize=12)
ax[0].grid(alpha=0.15)

# ---- Gráfica Lineweaver–Burke ----
# Ajuste lineal
m, b = np.polyfit(inv_S, inv_v, 1)

# Recta ajustada
x_line = np.linspace(min(inv_S), max(inv_S), 100)
y_line = m*x_line + b

ax[1].scatter(inv_S, inv_v, color="black", s=25)
ax[1].plot(x_line, y_line, color="red", linewidth=2)


ax[1].set_title("Lineweaver–Burke (doble recíproco)", fontsize=14)
ax[1].set_xlabel("1 / [S]", fontsize=12)
ax[1].set_ylabel("1 / v", fontsize=12)
ax[1].grid(alpha=0.15)

print(f"Vmax estimado = {1/b:.3f}")
print(f"Km estimado = {m/b:.3f}")

plt.tight_layout()
plt.show()
```

# 5. MECANISMOS DE DEGRADACIÓN CATALITICA 

La degradación catalítica es la perdida de actividad y selectividad del catalizador con el tiempo o bajo efectos de algunas sustancias, es un problema en los procesos industriales, ya que remplazar estas sustancias y parar el proceso es algo muy costoso. La degradación es muy diferente dependiendo al proceso que sea sometido; por ejemplo, en el craqueo la degradación puede ocurrir en segundos, mientras que en procesos como la síntesis de amoniaco el catalizador puede perdurar durante años.
La perdida de actividad es algo que se controla fuertemente, pero en ocasiones malos diseños o practicas afectan fuertemente los catalizadores. Algunos ejemplos de procesos que pueden favorecer la degradación de un catalizador pueden ser la temperatura, las malas relaciones o la exposición al ambiente.
Este proceso es totalmente inevitable para la mayoría catalizadores, pero algunas la velocidad o daño de las consecuencias se pueden evitar, posponer o revertir.
Hay muchos métodos en que se pueda dar la degradación catalítica. Por ejemplo, un catalizador solido puede ser contaminado por medio de su superficie o poros en reacciones de craqueo de reactivos. En tratamientos de gases de combustión un catalizador se puede ver contaminado, erosionado o taponado. En conclusión los catalizadores pueden verse afectados en procesos de degradación como: 






![Tomado de : https://www.scielo.org.mx/scielo.php?script=sci_arttext&pid=S2448-56912021000100005](degradación.png)

## 5.1. Saturación 
En sistemas donde los reactivos o productos se adsorben fuertemente, puede ocurrir saturación de la superficie: todos los sitios activos quedan ocupados, lo que limita la velocidad a una región de orden cero respecto a la concentración del reactivo.
Cuando especies fuertemente adsorbidas (por ejemplo, productos pesados o impurezas condensables) ocupan los poros y entradas, se habla de bloqueo. Tales como:

•	bloqueo intraporo (depósitos dentro de los poros que reducen su diámetro efectivo);

•	bloqueo de boca de poro (tapones en la entrada que impiden el acceso).

Ambos casos reducen la superficie accesible y la capacidad de transporte de masa dentro del catalizador.


![Tomado de: https://www.scielo.org.mx/scielo.php?script=sci_arttext&pid=S2448-56912021000100005](saturacion.png)

## 5.2. Envenenamiento 
El envenenamiento es la desactivación causada por adsorción casi irreversible de impurezas que neutralizan sitios activos. Revisiones clásicas y recientes identifican como venenos típicos al azufre, fósforo, metales pesados, halógenos y compuestos nitrogenados.
Los mecanismos principales son:

1.	Formación de complejos fuertemente adsorbidos (por ejemplo, sulfuros metálicos sobre catalizadores de hidrogenación).

2.	Modificación electrónica del metal activo, alterando su capacidad de adsorber y activar reactivos.

3.	Bloqueo estérico de sitios oclusos en el interior de poros.

En procesos de abatimiento de contaminantes atmosféricos (por ejemplo, reducción catalítica selectiva de NOx), se ha observado que el envenenamiento selectivo por ciertos compuestos de azufre reduce drásticamente la conversión a bajas temperaturas, lo que obliga a estrategias de pretratamiento del gas o diseño de catalizadores resistentes al veneno.

## 5.3. BLOQUEO 

En sistemas donde los reactivos o productos se adsorben fuertemente, puede ocurrir saturación de la superficie: todos los sitios activos quedan ocupados, lo que limita la velocidad a una región de orden cero respecto a la concentración del reactivo.
Cuando especies fuertemente adsorbidas (por ejemplo, productos pesados o impurezas condensables) ocupan los poros y entradas, se habla de bloqueo. Tales como:

•   bloqueo intraporo (depósitos dentro de los poros que reducen su diámetro efectivo);

•   bloqueo de boca de poro (tapones en la entrada que impiden el acceso).

Ambos casos reducen la superficie accesible y la capacidad de transporte de masa dentro del catalizador.

En muchas reacciones de hidrocarburos, la principal causa de desactivación es la formación de coque: depósitos carbonosos que se forman por reacciones de polimerización y condensación de intermediarios insaturados.

![Tomado de: https://www.samaterials.es/how-to-prevent-catalyst-deactivation.html](bloqueo.JPg)

## 5.4. Sinterización 
La sinterización es un proceso térmico en el que partículas metálicas o del soporte se reagrupan formando agregados mayores, disminuyendo la superficie específica y el número de sitios activos. Revisiones de sinterización y estabilidad de catalizadores muestran que este fenómeno está promovido por altas temperaturas, atmósferas reductoras u oxidantes agresivas y ciclos térmicos repetidos.
A nivel microscópico, la sinterización puede darse por:

•	migración y coalescencia de partículas metálicas sobre la superficie del soporte;

•	disolución–reprecipitación , donde átomos se disuelven y vuelven a depositarse en partículas grandes.

## 5.5. Reacciones secundarias y vida útil del catalizador
Las reacciones secundarias, a menudo favorecidas por altas temperaturas o tiempos de residencia largos, generan subproductos pesados o altamente insaturados que contribuyen a coquización, envenenamiento o bloqueo de poros.
Modelos de desactivación consideran que la velocidad de pérdida de actividad está relacionada con la formación de estos subproductos, de modo que condiciones que minimicen las reacciones paralelas (por ejemplo, control cuidadoso de temperatura y estequiometría de alimentación) prolongan la vida útil del catalizador.  

## 5.6. Estrategias para prevenir o minimizar la degradación catalítica
Las revisiones recientes enfatizan estrategias integradas que combinan diseño de catalizador y control de proceso:


### 5.6.1.	Diseño de catalizadores más robustos

•	uso de soportes térmicamente estables (por ejemplo, óxidos mixtos);

•	dopado con promotores que mejoren resistencia a sinterización y envenenamiento;

•	introducción de mesoporos para reducir bloqueo y difusiones lentas.


### 5.6.2.	Control de la composición de alimentación

•	pretratamiento para remover azufre, cloruros o metales pesados;

•	eliminación de oxígeno o agua en procesos donde promuevan degradación del soporte.


### 5.6.3.	Optimización de condiciones de operación

•	limitar temperaturas máximas para evitar sinterización;

•	operar en rangos de conversión que minimicen reacciones secundarias formadoras de coque;

•	diseño hidrodinámico que reduzca zonas muertas y gradientes de temperatura.MDPI+1


### 5.6.4.	Estrategias de regeneración programada

•	ciclos periódicos de regeneración in situ (por ejemplo, combustión de coque en reactores de lecho fijo o móvil);

•	operación en reactores de lecho fluidizado o de lecho móvil continuo que permitan regeneración simultánea.



![Tomado de: https://www.samaterials.es/how-to-prevent-catalyst-deactivation.html](images5.JPg)

## 5.7. Mecanismos de regeneración de catalizadores
Dependiendo del mecanismo de desactivación, es posible recuperar parcial o totalmente la actividad mediante regeneración:
Para coquización: combustión controlada del carbono con aire o vapor (gasificación).

•	Para envenenamiento reversible: desorción térmica o con atmósferas reductoras/oxidantes adecuadas.

•	Para sinterización: en general es irreversible; sólo se corrige reemplazando el catalizador.

•	Para cambios de fase: tratamientos térmicos o químicos específicos que restauren la fase activa (cuando es posible).

### 6.1.1. APLICACIONES INDUSTRIALES DE LA CATALISIS HOMOGENEA

 En la catálisis homogénea se puede encontrar una de las partes más importantes dentro de los procesos industriales modernos, específicamente en aquellos de selectividad fina y el control molecular, como pasa en la producción farmacéutica o en la producción de químicos de alto valor.  En estas aplicaciones la catálisis homogénea transforma sustratos complejos con mucha precisión, dando mejoras en la eficiencia, en la reducción de subproductos y pasos de purificación.
 A nivel global la cantidad de procesos industriales que dependen de la catálisis homogénea es enorme, muchos de estos basados en complejos metálicos, como en la hidroformilación, que permite la formación de aldehídos a gran escala y con una altísima selectividad, y esto a su vez, impacta la producción de detergentes, plásticos y solventes. Las reacciones de polimerización controlada gracias a la catálisis se han podido abrir rutas industriales que eran consideradas, prácticamente inviables.
 En la selección en planta, el catalizador homogéneo requerido se escoge respecto a su estabilidad, su tolerancia a impurezas y su capacidad de recuperarse o reciclarse, siendo esta su facultad más importante, ya que el costo de los metales de transición utilizados es muy alto. Por eso en la actualidad se busca nuevas estrategias como lo son los disolventes ionios, membranas selectivas o sistemas bifásicos acuosos que permitan acuosos, que permitan conservar la selectividad de la catálisis homogénea, pero con costos de utilización rentables. Estos avances científicos en los últimos años han permitido usar de manera industrial la catálisis homogénea que es esencial para la química contemporánea.

### 6.1.2.  APLICACIONES INDUSTRIALES DE LA CATALISIS HETEROGENEA 

La catálisis heterogénea es prácticamente uno de los motores fundamentales en toda la industria química actual. Estos procesos son utilizados ampliamente en refinerías, plantas petroquímicas y unidades de conversión energética, apareciendo en el día a día de las personas, ya sea en combustibles, fertilizantes, plásticos, o incluso materiales muchísimo mas avanzados. 
Sim la catálisis heterogénea la gran mayoría de productos serían imposible de hacer de una manera viable ya que todos estos dependen de los catalizadores sólidos. El fin de usar este tipo de catalizadores es reducir los costos operativos y el costo energético. Por ejemplo, en el reformado catalítico y el craqueo en refinerías, los catalizadores solidos permiten transformar grandes cantidades de petróleo en sus derivados de alto valor.
En la selección del catalizador se debe tener en cuenta que el catalizador no solo debe ser funcional al inicio del proceso, sino que pueda soportar gran cantidad de horas sin desactivarse, por ende, actualmente se busca la estabilidad, la dispersión metálica controlada y resistencia a la sinterización o al envenenamiento de impurezas. Este tipo de catálisis es la base de energías emergentes como la transformación de la biomasa la utilización del CO₂ y la producción de hidrogeno. Todos estos factores convierten la catálisis heterogénea en la base de la industria actual y del futuro.
